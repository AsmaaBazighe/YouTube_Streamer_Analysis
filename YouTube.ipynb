# Performance Metrics:
- Calculate and visualize the average number of subscribers, visits, likes, and comments.
- Are there patterns or anomalies in these metrics?

```python
# Calculate the average number of subscribers, visits, likes, and comments
avg_subscribers = df['Suscribers'].mean()
avg_visits = df['Visits'].mean()
avg_likes = df['Likes'].mean()
avg_comments = df['Comments'].mean()

# Print the average metrics
print("Average Subscribers:", avg_subscribers)
print("Average Visits:", avg_visits)
print("Average Likes:", avg_likes)
print("Average Comments:", avg_comments)
```

- **Average Subscribers**: The average number of subscribers among streamers is approximately 22,415,561.96. This metric indicates the overall popularity of the streamers, with higher subscriber counts suggesting greater influence or appeal.
- **Average Visits**: The average number of visits for streamers is around 1,210,729.68. This metric represents the average traffic or viewership each streamer receives. Higher visit counts may suggest higher engagement or visibility of the content.
- **Average Likes**: The average number of likes per streamer is about 53,473.60. Likes are a measure of audience engagement and satisfaction with the content. Higher average likes indicate a positive reception of the content by the audience.
- **Average Comments**: The average number of comments per streamer is approximately 1,558.79. Comments reflect audience interaction and feedback on the content. Higher average comments may indicate active engagement and discussion around the streamer's content.

```python
# Visualize the average metrics
metrics = ['Subscribers', 'Visits', 'Likes', 'Comments']
avg_values = [avg_subscribers, avg_visits, avg_likes, avg_comments]

plt.figure(figsize=(10, 6))
plt.bar(metrics, avg_values, color='skyblue')
plt.title('Average Performance Metrics for YouTube Streamers')
plt.xlabel('Metrics')
plt.ylabel('Average Value')
plt.show()
```

- **Subscribers vs. Visits**:
High average subscribers coupled with low average visits may indicate that although the streamers have a large subscriber base, their content may not be effectively reaching or engaging a significant portion of their audience. This could suggest potential issues with content promotion or audience retention.


- **Likes vs. Comments**:
The average likes significantly outnumber the average comments, it may indicate that the audience prefers passive engagement by liking the content rather than actively commenting. 


- **Subscribers vs. Likes vs. Comments**:
Observing the relationship between the average subscribers, likes, and comments can provide a comprehensive view of audience engagement. A discrepancy between the average subscriber count and the average likes or comments may indicate that the streamers have a large subscriber base but low engagement levels, suggesting potential issues with content quality or audience targeting.


- **Visits vs. Likes vs. Comments**:
Analyzing the average visits alongside the average likes and comments can highlight the effectiveness of content in driving audience interaction. A high average visit count coupled with low average likes or comments may indicate that the streamers attract a large audience but struggle to engage them effectively.

# Content Recommendations:
- Propose a system for enhancing content recommendations to YouTube users based on streamers categories and performance metrics.

```python
#use only necessary columns
data = df[['Username', 'Categories', 'Suscribers', 'Visits', 'Likes', 'Comments']]
# Encode categorical variables 
data = pd.merge(data, categories_df, left_on='Categories', right_on='Categories', how='left')
data.drop(['Categories', 'Categories'], axis=1, inplace=True)
data
```

```python
# Calculate engagement rate
data['EngagementRate'] = (data['Likes'] + data['Comments']) / data['Visits']
data
```

```python
data.isnull().sum()
```

```python
# Handle missing values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Standardize numerical features
scaler = StandardScaler()
numerical_cols = ['Suscribers', 'Visits', 'Likes', 'Comments', 'category_ID', 'EngagementRate']
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Fit k-NN model
knn_model = NearestNeighbors(n_neighbors=4, metric='euclidean')
knn_model.fit(data[numerical_cols])

def recommend_similar_streamers(target_streamer_name):
    # Find index of target streamer
    if target_streamer_name in data['Username'].values:
        target_streamer_index = data[data['Username'] == target_streamer_name].index[0]
    
        # Query k-NN model to find similar streamers
        target_data = data.loc[[target_streamer_index], numerical_cols]  # Create DataFrame with single row
        distances, indices = knn_model.kneighbors(target_data)
    
        # Get names of similar streamers
        similar_streamers = data.loc[indices[0], 'Username'].tolist()
    
        # Remove target streamer from the list (if present)
        if target_streamer_name in similar_streamers:
            similar_streamers.remove(target_streamer_name)
    
        # Return recommended streamers
        return similar_streamers[:3]  # Return top 3 similar streamers
    else:
        print("We are sorry, but we don't have data about this streamer, so we can't presently provide you with recommendations.")

```

```python
# Example usage:
target_streamer_name = input("Enter the name of your the target streamer: ")
recommended_streamers = recommend_similar_streamers(target_streamer_name)
print("Recommended streamers for", target_streamer_name, ":", recommended_streamers)
```
